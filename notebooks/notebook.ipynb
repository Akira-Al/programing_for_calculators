```python
def polar2score(p: str) -> int:
    match p:
        case "p":
            return 1
        case "n":
            return -1
        case "e":
            return 0
        case _:
            raise ValueError(f"Invalid polar: {p}")
```

```python
def evaluate(text: str, reference: dict) -> int:
    score = 0
    for token in tokenizer.tokenize(text):
        if token.surface in reference:
            s = polar2score(reference[token.surface][0])
            print(f"{token.surface}: {s}")
            score +=  s
    return score
```

```python
def partial_match(tokens: list, reference: list) -> bool:
    for i in range(len(tokens)):
        if tokens[i] != reference[i]:
            return False
    return True
```

```python
def evaluate2(text: str) -> int:
    score = 0
    is_pending = False
    pending = []
    for token in tokenizer.tokenize(text):
        if token.surface in reference1_dict:
            s = polar2score(reference1_dict[token.surface][0])
            print(f"[ref1] {token.surface}: {s}")
            score +=  s
        if token.surface in reference2_dict:
            is_pending = True

        if is_pending:
            pending.append(token.surface)
            has_partial_match = False
            for record in reference2_dict[pending[0]]:
                if pending == record[0]:
                    s = record[1]
                    print(f"[ref2] {pending}: {s}")
                    score += s
                    pending = []
                    is_pending = False
                    break
                elif partial_match(pending, record[0]):
                    has_partial_match = True
            else:
                if not has_partial_match:
                    is_pending = False
                    pending = []
                    
        
        
    return score
```

```python
with open("../data/data_ans.txt", "r", encoding="UTF-8") as f:
    data_text_with_ans = f.readlines()
l_pol_and_text=[]
for l in data_text_with_ans:
    pol_and_text=l.split(" ")
    pol_and_text[1]=pol_and_text[1].replace("\n","")
    
    l_pol_and_text.append(pol_and_text)
def pol_to_sign(pol:str)->int:
    if(pol=="e"):
        return 0
    if(pol=="n"):
        return -1
    if(pol=="p"):
        return 1
    return None
l_pol_and_text
```

```python
def evaluate2_verbose(text: str) -> {int,str}:
    debug_output=""
    score = 0
    is_pending = False
    pending = []
    for token in tokenizer.tokenize(text):
        debug_output+=str(token)+"\n"
        if token.surface in reference1_dict:
            s = polar2score(reference1_dict[token.surface][0])
            debug_output+="[ref1] "+str(token.surface)+": "+str(s)+"\n"
            score +=  s
        if token.surface in reference2_dict:
            is_pending = True

        if is_pending:
            pending.append(token.surface)
            has_partial_match = False
            for record in reference2_dict[pending[0]]:
                if pending == record[0]:
                    s = record[1]
                    debug_output+="[ref2] "+str(pending)+": "+str(s)+"\n"
                    score += s
                    pending = []
                    is_pending = False
                    break
                elif partial_match(pending, record[0]):
                    has_partial_match = True
            else:
                if not has_partial_match:
                    is_pending = False
                    pending = []
                    
        
        
    return score,debug_output
```

```python

evaluate2_verbose(l_pol_and_text[3][1])
```

```python
count_diff = 0
for t in l_pol_and_text:
    score, text = evaluate2_verbose(t[1])
    pol_sign = pol_to_sign(t[0])
    success = True
    if(pol_sign==0):
        if(score!=0):
            success=False
    elif(int(pol_sign)*int(score)<=0):
        success=False
    if not success:
        count_diff+=1
        print("↓")
        print(t)
        print("======")
        print("evaluated score:",score,", ans:",pol_sign)
        print(text)
        print("↑")
print("sum of number  of diffs:",count_diff)
```
